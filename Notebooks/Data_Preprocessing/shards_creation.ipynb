## üìä Load Training CSV and Normalize Latitude & Longitude üåç


```python
new_df_train = pd.read_csv("new_df_train.csv")
```

```python
new_df_train
```

```python
lat_mean, lat_std = new_df_train["latitude"].mean(), new_df_train["latitude"].std()
lon_mean, lon_std = new_df_train["longitude"].mean(), new_df_train["longitude"].std()
```

```python
new_df_train["latitude_norm"]  = (new_df_train["latitude"] - lat_mean) / lat_std
new_df_train["longitude_norm"] = (new_df_train["longitude"] - lon_mean) / lon_std
```

## üóÇÔ∏è Generate Training Shards with Images, Labels, and Normalized Coordinates üåç


```python
import os, io, tarfile
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm   # progress bar

root_path = r"C:\Users\thaku\jupyter_notebook _datasets\Wildlife_dataset"

def create_shards(output_dir, df, samples_per_shard=1000):
    os.makedirs(output_dir, exist_ok=True)

    for shard_id, start in enumerate(range(0, len(df), samples_per_shard)):
        end = min(start + samples_per_shard, len(df))
        shard_df = df.iloc[start:end]

        shard_path = os.path.join(output_dir, f"shard-{shard_id:05d}.tar")
        with tarfile.open(shard_path, "w") as tar:
            # tqdm loop for samples
            for idx, row in tqdm(
                shard_df.iterrows(),
                total=len(shard_df),
                desc=f"Shard {shard_id:05d}",
                unit="sample"
            ):
                key = f"{idx:06d}"   # unique key

                # --- image ---
                img_path = os.path.join(root_path, row["img_path"])
                tar.add(img_path, arcname=f"{key}.jpg")

                # --- class ---
                label_bytes = str(row["class"]).encode("utf-8")
                info = tarfile.TarInfo(f"{key}.cls")
                info.size = len(label_bytes)
                tar.addfile(info, io.BytesIO(label_bytes))

                # --- latitude ---
                if "latitude" in row and not pd.isna(row["latitude"]):
                    lat_bytes = str(row["latitude_norm"]).encode("utf-8")
                    info = tarfile.TarInfo(f"{key}.lat")
                    info.size = len(lat_bytes)
                    tar.addfile(info, io.BytesIO(lat_bytes))

                # --- longitude ---
                if "longitude" in row and not pd.isna(row["longitude_norm"]):
                    lon_bytes = str(row["longitude"]).encode("utf-8")
                    info = tarfile.TarInfo(f"{key}.lon")
                    info.size = len(lon_bytes)
                    tar.addfile(info, io.BytesIO(lon_bytes))

        print(f"‚úÖ Shard saved: {shard_path}")

create_shards(
    output_dir=r"C:\Users\thaku\jupyter_notebook _datasets\Wildlife_dataset\shards_train_mini_300",
    df=new_df_train,
    samples_per_shard=4500
)
```
